```postgresql
CREATE OR REPLACE FUNCTION expon(N INTEGER, l FLOAT = 1)
RETURNS FLOAT[] AS
$BODY$
SELECT
    array_agg(-(1/l)*ln(1 - random()::FLOAT))
FROM
    generate_series(1, N) AS i;
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT expon(100, 0.5);
```

```postgresql
CREATE OR REPLACE FUNCTION norm(N INTEGER, mu FLOAT = 0, sigma FLOAT = 1)
RETURNS FLOAT[] AS
$BODY$
SELECT
    array_agg(sigma*sqrt(-2.*ln(random()::FLOAT))*cos(4*acos(0.)*random()::FLOAT)+mu)
FROM
    generate_series(1, N) AS i;
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT norm(100);
```

```postgresql
CREATE OR REPLACE FUNCTION lognorm(N INTEGER, mu FLOAT = 0, sigma FLOAT = 1)
RETURNS FLOAT[] AS
$BODY$
SELECT
    array_agg(exp(x))
FROM
    unnest(norm(N, mu, sigma)) AS x;
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT lognorm(100);
```

```postgresql
CREATE OR REPLACE FUNCTION f(x FLOAT, args FLOAT[] = '{2,3}'::FLOAT[])
RETURNS FLOAT AS
$BODY$
SELECT args[1]*x+args[2];
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
DROP FUNCTION IF EXISTS intdiff;
CREATE OR REPLACE FUNCTION intdiff(
    f     regprocedure,
    args  FLOAT[] = NULL::FLOAT[],
    xmin  NUMERIC = -1.,
    xmax  NUMERIC = 10,
    dx    NUMERIC = 0.1
) RETURNS TABLE(
    x     FLOAT,
    fx    FLOAT,
    dx    FLOAT,
    dfdx  FLOAT,
    fxdx  FLOAT,
    Ifxdx FLOAT
) AS
$BODY$
WITH 
D AS (
SELECT
    x::FLOAT,
    f(x::FLOAT, args) AS fx,
    dx::FLOAT,
    LEAD(f(x::FLOAT, args)) OVER(ORDER BY x) AS fx1
FROM generate_series(xmin, xmax, dx) AS x
),
D2 AS (
SELECT
    D.x, D.fx, D.dx, 
    (D.fx1 - D.fx)/dx AS dfdx,
    (D.fx1 + D.fx)*D.dx/2 AS mfxdx
FROM D
)
SELECT
    D2.*,
    SUM(D2.mfxdx) OVER (ORDER BY x ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW)
FROM D2
ORDER BY D2.x;
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
DROP FUNCTION IF EXISTS trapz;
CREATE OR REPLACE FUNCTION trapz(
    _func regprocedure,
    args  FLOAT[] = NULL::FLOAT[],
    xmin  NUMERIC = -1.,
    xmax  NUMERIC = 10,
    dx    NUMERIC = 0.1
) RETURNS FLOAT AS
$BODY$
SELECT SUM(fxdx) FROM intdiff(_func, args, xmin, xmax, dx)
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT * FROM intdiff('f(FLOAT, FLOAT[])'::regprocedure, '{3,4}'::FLOAT[])
```

```postgresql
SELECT * FROM trapz('f(FLOAT, FLOAT[])', '{3,4}'::FLOAT[])
```

```postgresql
CREATE OR REPLACE FUNCTION normexp(x FLOAT, args FLOAT[] = NULL::FLOAT[])
RETURNS FLOAT AS
$BODY$
SELECT exp(-x^2);
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
CREATE OR REPLACE FUNCTION erf(x FLOAT, dx NUMERIC = 0.1)
RETURNS FLOAT AS
$BODY$
SELECT trapz('normexp(FLOAT, FLOAT[])', '{}', 0., x::NUMERIC, dx);
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT normexp(-1)
```

```postgresql
CREATE OR REPLACE FUNCTION erf(x FLOAT, dx NUMERIC = 1e-3)
RETURNS FLOAT AS
$BODY$
WITH
D AS (
SELECT
    y::FLOAT,
    exp(-((y::FLOAT)^2)) AS fx0,
    LEAD(exp(-((y::FLOAT)^2))) OVER(ORDER BY y) AS fx1
FROM generate_series(0, x::NUMERIC, dx) AS y)
SELECT
    COALESCE((2/sqrt(2*acos(0.)))*SUM((D.fx1 + D.fx0)*dx::FLOAT/2), 0.)
FROM D;
$BODY$
LANGUAGE SQL IMMUTABLE;
```

```postgresql
SELECT
    x::FLOAT
   ,erf(x::FLOAT)
FROM
    generate_series(0, 3.0, 0.1) AS x
```
