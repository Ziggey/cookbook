```python
sensor_type = ('sds011','bme280','dht22')
start = date(2020, 2, 10)
end = date(2020, 3, 8)
```

```python
# https://stackoverflow.com/a/1060330/12463368
def daterange(start, end):
    for n in range(int ((end - start).days)):
        yield start + timedelta(n)

dt = []
for single_date in daterange(start, end):
    d = single_date.strftime("%Y-%m-%d")
    dt.append(d)
```

## Data Extraction

```python
l = []

for i in info.sensor:
    for j in dt:
        for s in sensor_type:
            file = '{0}_{1}_sensor_{2}.csv'.format(j,s,i)
            url = 'http://archive.luftdaten.info/{0}/{1}'.format(j,file)
            r = requests.get(url)
            c = r.content
            print(r, c[:100])
            df = pd.read_csv(io.StringIO(c.decode('utf-8')),sep=';')
            # save csv!!!!
            l.append(df)
    
frame = pd.concat(l, axis=0, ignore_index=True, sort = True)
```

```python
frame.tail()
```

```python
df = frame[['timestamp','sensor_id','sensor_type','P1','P2']]
```

```python
df.loc[:,'timestamp'] = pd.to_datetime(df.timestamp)
```

```python
df.set_index('timestamp',inplace=True)
```

```python
df.loc[:,'sensor_id'] = df['sensor_id'].fillna(0).astype(int)
```
