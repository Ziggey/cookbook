```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
pd.__version__
```

```python
t = pd.date_range('1990-01-01', '1990-01-08', freq='1H')
#t = pd.date_range('1990-01-01', '1990-01-08', freq='1H', origin='unix', units='D')
t
```

```python
t[-1] - t[0]
```

```python
x = pd.DataFrame(np.random.rand(len(t)), index=t)
x.iloc[:10]
```

```python
axe = x.plot()
axe.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
axe.xaxis.set_minor_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
plt.setp(axe.xaxis.get_majorticklabels(), rotation=90)
plt.setp(axe.xaxis.get_minorticklabels(), rotation=90)
plt.show(axe)
```

```python
pandasTicks = axe.get_xticks()
pandasTicks
```

```python
fig, axe = plt.subplots()
axe.plot(x)
axe.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
plt.setp(axe.xaxis.get_majorticklabels(), rotation=90)
plt.show(axe)
```

```python
MatplotlibTikcs = axe.get_xticks()
MatplotlibTikcs
```

```python
ticks = pd.date_range('1990-01-01', '1990-01-08', freq='6H')
axe = x.plot(xticks=ticks)
axe.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
plt.setp(axe.xaxis.get_majorticklabels(), rotation=90)
plt.show(axe)
```

```python
#ticks = x.resample('6H').index
ticks = pd.date_range('1990-01-01', '1990-01-08', freq='6H')
axe = x.plot(xticks=ticks)
axe.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
axe.set_xticklabels(ticks.strftime("%a %H:%M"))
axe.set_xticklabels([], minor=True)
plt.setp(axe.xaxis.get_majorticklabels(), rotation=90)
plt.show(axe)
```

```python
fig, axe = plt.subplots()
axe.plot(x)
axe.xaxis.set_minor_locator(mdates.DayLocator())
axe.xaxis.set_minor_locator(mdates.HourLocator(byhour=(6,12,18)))
axe.xaxis.set_major_formatter(mdates.DateFormatter("%A"))
axe.xaxis.set_minor_formatter(mdates.DateFormatter("%H:%M"))
plt.setp(axe.xaxis.get_majorticklabels(), rotation=90)
plt.setp(axe.xaxis.get_minorticklabels(), rotation=90)
axe.xaxis.grid(which='major', color='black', linewidth=1, linestyle='-')
axe.xaxis.grid(which='minor', color='black', linewidth=1, linestyle=':')
axe.yaxis.grid()
plt.show(axe)
```
