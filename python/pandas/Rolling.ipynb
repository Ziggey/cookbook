```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
n = 25
t = np.arange(n)
x = (3*np.exp(0.2*t)).astype(int)
y = (0.3*t).astype(int)
```

```python
def m(x):
    return x['y'].sum()/x['x'].sum()
```

```python
df = pd.DataFrame(np.array([x, y]).T, index=t, columns=['x', 'y'])
df
```

```python
df.rolling(7).apply(m, raw=False)
```
