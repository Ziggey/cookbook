```python
import numpy as np
from scipy import stats, signal, fftpack
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 6000
t = np.linspace(0, 600, N)
```

```python
xt = stats.norm(loc=8, scale=2).pdf(t[:200])
xt = xt/np.sum(xt)
```

```python
gt = stats.lognorm(0.25, loc=-20, scale=100).pdf(t[:3000])
gt = gt/np.sum(gt)
```

```python
yt = signal.convolve(xt, gt, mode='valid')
```

```python
xt = np.pad(xt, (0, N-xt.size), 'constant')
gt = np.pad(gt, (0, N-gt.size), 'constant')
yt = np.pad(yt, (0, N-yt.size), 'constant')
```

```python
fig, axe = plt.subplots()
axe.plot(t, xt, label="$x(t)$")
axe.plot(t, gt, label="$g(t)$")
axe.plot(t, yt, label="$y(t)$")
#axe.set_xlim([15,35])
axe.legend()
axe.grid()
```

```python
g = fftpack.fft(gt)
```

```python
fig, axe = plt.subplots()
axe.semilogy(np.real(g), label="$X(s)$")
```
