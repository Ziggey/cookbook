```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi, 1001)
y = 3 + 2*np.sin(12*x)
```

```python
fig = plt.figure()

axe1 = fig.add_axes((0.2,0.2,0.8,0.8))
axe1.plot(x, y)

axe2 = fig.add_axes((0.1,0.1,0.2,0.2))
img=mpimg.imread('bug.png')
imgplot = axe2.imshow(img)
```
